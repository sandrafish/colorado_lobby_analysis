```python
import pandas as pd
import numpy as np
```

```python
#subcontractor table
subcon = pd.read_csv('Discl_con.txt', sep='\t', names = ['lob_last', 'lob_first', 'lob_add1', 'lob_add2', 'lob_city', 'lob_state',
                                                        'lob_zip', 'lob_id', 'lob_annual_id', 'income_source', 'sub_add1',
                                                         'sub_add2', 'sub_city', 'sub_state', 'sub_zip', 'income_rec', 'paid_to_sub',
                                                          'receipt_date', 'sub_last', 'sub_first', 'report_month',
                                                        'fiscal_year', 'due_date'], 
                                                dtype={'lob_id': 'str', 
                                                 'lob_annual_id': 'str',
                                            
                                                 'fiscal_year': 'str',
                                                      })
subcon.info()
```

First, taking the commas out of numbers.

```python
# convert amount to float
subcon['income_rec'] = subcon['income_rec'].str.replace(',', '')
subcon.head(10)
```

```python
# convert amount to float
subcon['paid_to_sub'] = subcon['paid_to_sub'].str.replace(',', '')
subcon.head(10)
```

Next, combining lobbyist and subcontractor names.

```python
#combine lobbyist name - if the name is not combined, you can't get correct sums
subcon['lobname'] = subcon['lob_first'].fillna('') + ' ' + subcon['lob_last'].fillna('')
#strip leading space
subcon['lobname'] = subcon['lobname'].str.strip()
subcon.head(25)
```

```python
#combine lobbyist name - if the name is not combined, you can't get correct sums
subcon['subcon_name'] = subcon['sub_first'].fillna('') + ' ' + subcon['sub_last'].fillna('')
#strip leading space
subcon['subcon_name'] = subcon['subcon_name'].str.strip()
subcon.head(25)
```

Getting rid of uppercase stuff.

```python
detail = subcon.apply(lambda x: x.astype(str).str.title())
detail.info()
detail.head()
```

Converting income from string to numbers.

```python
detail['income_rec'] = pd.to_numeric(detail['income_rec'], errors='coerce')
detail['paid_to_sub'] = pd.to_numeric(detail['paid_to_sub'], errors='coerce')
detail.info()
```

There are some records here that are also reported by lobbyists as payments in the income files, i.e., duplication. The next three cells get rid of that duplication in the subcontractor file.

```python
to_drop = ['Richard Orf', 'Ted Orf', 'Dianna Orf', 'Alexandra Garza', 'Colorado Advocates', 'Andrew Struttmann']
subcon17 = subcon17[~subcon17['subcon_cleanname'].isin(to_drop)]
subcon17.info()
```

```python
subcon17 = subcon17[~((subcon17['lobbyist_cleanname'] == 'Ontheballot Consulting') & (subcon17['subcon_cleanname'] == 'Brett Moore'))]
subcon17.info()
```

```python
subcon17 = subcon17[~((subcon17['lobbyist_cleanname'] == 'Steven Delbianco') & (subcon17['paid_to_sub'] == 5000))]
subcon17.info()
```
