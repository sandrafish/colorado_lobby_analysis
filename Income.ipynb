Import the income text file from the Colorado Secretary of State. Some of these items need to be forced into string data types.

```python
df = pd.read_csv('Discl_inc.txt', sep='\t', names = ['lob_last', 'lob_first', 'lob_add1', 'lob_add2', 'lob_city', 'lob_state',
                                                        'lob_zip', 'lob_id', 'lob_annual_id', 'emp_name', 'emp_last', 'emp_first',
                                                       'employer_mid', 'employer_add_1', 'employer_add_2', 'employer_cit',
                                                        'employer_state', 'employer_zip', 'biz_descip', 'biz_type', 'emp_status', 
                                                        'emp_start_date', 'emp_end_date', 'income', 'income_received_date', 'report_month',
                                                        'fiscal_year', 'due_date'], 
                                                dtype={'lob_id': 'str', 
                                                 'lob_annual_id': 'str',
                                                       'employer_mid': 'str',
                                                 'fiscal_year': 'str',
                                                      })
df.info()
```

```python
df.head()
```

Next is the cleanup process. First combining the lobbyists names, which isn't perfect because things like 5280 Strategies become Michael 5280 Strategies. Then the employer name.

```python
#combine lobbyist name - if the name is not combined, you can't get correct sums
df['lobname'] = df['lob_first'].fillna('') + ' ' + df['lob_last'].fillna('')
#strip leading space
df['lobname'] = df['lobname'].str.strip()
df.head(25)
```

```python
#combine employer name - if the name is not combined, you can't get correct sums
df['empname'] = df['emp_name'].fillna('') + df['emp_first'].fillna('') + ' ' + df['emp_last'].fillna('')
#strip leading space
df['empname'] = df['empname'].str.strip()
df.head(25)
```

This changes the case to title case, eliminating all caps. This can create problems - it eliminates the NaN settings (and there's probably a way to avoid this that i should research), and messes up company names that are all caps: AT&T becomes At&T for instance.

```python
detail = df.apply(lambda x: x.astype(str).str.title())
detail.info()
detail.head()
```

Then we need to make the income field a number, which is a two step process. First, removing the commas, then coercing it into a float.

```python
# convert amount to float
detail['income'] = detail['income'].str.replace(',', '')
detail.head(10)
```

```python
detail['income'] = pd.to_numeric(detail['income'], errors='coerce')
detail.head(10)
detail.info()
```
